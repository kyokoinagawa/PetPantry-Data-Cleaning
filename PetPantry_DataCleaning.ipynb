```python
df = pd.read_csv('FY23 Split Animal Table.csv')
df.head()
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = ["Animal Names", "Breed", "Age", "Weight"])
```

```python
df
```

```python
df["Phone Number"] =df["Phone Number"].apply(lambda x: str(x))
```

```python
df["Phone Number"] = df["Phone Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone Number"] = df["Phone Number"].str.replace('nan--', '')
df["Phone Number"] = df["Phone Number"].str.replace('Na--', '')
df
```

```python
df.to_csv('FY23 Cleaned Phone Table.csv')
```

```python
weight_cols = [f'Weight_{i}' for i in range(1, 12)]
breed_cols = [f'Breed_{i}' for i in range(1, 12)]

# Clean weight columns
for col in weight_cols:
    df[col] = df[col].str.extract('(\d+) lb')[0]

# Clean breed columns
for col in breed_cols:
    df[col] = df[col].str.extract('(\w+)$')[0]

print(df)
```

```python
age_cols = [f'Age_{i}' for i in range(1, 12)]

# Loop through each age column
for col in age_cols:
    # Extract the numeric age value and convert it to float for calculations
    df[f'{col}_value'] = df[col].str.extract('(\d+)')[0].astype(float)

    # Extract the unit (e.g., "yo", "months", "years")
    df[f'{col}_unit'] = df[col].str.extract('([a-zA-Z]+)[\s]*$')[0]

    # Update age values in 'years' where the unit is 'months'
    df.loc[df[f'{col}_unit'].str.contains('months', case=False, na=False), f'{col}_value'] /= 12.0

    # Drop the temporary unit column
    df.drop(columns=[f'{col}_unit'], inplace=True)

    # Replace the original age column with the updated values in 'years'
    df[col] = df[f'{col}_value']

    # Drop the temporary value column
    df.drop(columns=[f'{col}_value'], inplace=True)

# Display the updated DataFrame
print(df)
```

```python
df.to_csv('FY23 Cleaned COlumns Table.csv')
```

```python
df = pd.read_csv('FY23 Cleaned Phone Table2.csv')
df.head()
```

```python
weight_cols = [f'Weight_{i}' for i in range(1, 12)]
breed_cols = [f'Breed_{i}' for i in range(1, 12)]
name_cols = [f'Animal Name_{i}' for i in range(1, 12)]

# Clean weight columns
for col in weight_cols:
    df[col] = df[col].str.extract('(\d+) lb')[0]

# Clean breed columns
for col in breed_cols:
    df[col] = df[col].str.extract('(\w+)$')[0]

# Clean animal name columns
for col in name_cols:
    df[col] = df[col].str.extract('(\w+)$')[0]

print(df)
```

```python
df.to_csv('FY23 Cleaned Columns Table Except Age.csv')
```

```python
age_cols = [f'Age_{i}' for i in range(1, 12)]

def convert_to_years(age_str):
    if pd.isna(age_str) or age_str.lower() == 'unknown':
        return None  # or another value to indicate an unknown age

    try:
        age_value = float(age_str.split()[1])
        unit = age_str.split()[2].lower()
    except (ValueError, IndexError):
        return None  # or another value to indicate an error

    if 'y' in unit:
        return age_value
    elif 'm' in unit:
        return age_value / 12.0
    elif 'w' in unit:
        return age_value / 52.1775  # More accurate conversion to years

    return age_value  # Fallback to just return the age value as-is


# Loop through each age column and convert to years
for col in age_cols:
    df[col] = df[col].apply(convert_to_years)

# Display the updated DataFrame
print(df)
```

```python
df.to_csv('FY23 Cleaned Columns Table With Age.csv')
```
